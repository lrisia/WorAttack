```python
from nltk.corpus import wordnet
from nltk.corpus import words
from random import choice

amount = [5000, 4000, 3000, 2000, 1000]
while sum(amount) > 0:
    word = choice(words.words())
    try:
        syns = wordnet.synsets(word)
        meaning = syns[0].definition()
        print(f"{word}: {syns[0].definition()}")
        if 3 <= len(word) <= 4: amount[0] -= 1
        elif 4 <= len(word) <= 5: amount[1] -= 1
        elif 5 <= len(word) <= 6: amount[2] -= 1
        elif 6 <= len(word) <= 7: amount[3] -= 1
        elif 7 <= len(word) <= 10: amount[4] -= 1
        else: continue
        f = open("/content/drive/MyDrive/Colab Notebooks/WorAttack/word_nltk.txt", 'a')
        f.write(f"{word}\|\|{meaning}\n")
        f.close()
    except:
        pass
```
